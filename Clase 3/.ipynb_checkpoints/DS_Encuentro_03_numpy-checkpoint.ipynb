### Memoria y Velocidad

```python
import time
```

```python
%%time
L1 = range(100000)
L2 = range(100000)

A1 = np.arange(100000)
A2 = np.arange(100000)

s = time.time()
%time [a1+a2 for a1, a2 in zip(L1,L2)]
#print('python list:', time.time()-s)


s = time.time()
%time [L1[i]+L2[i] for i in range(len(L1))]
#print('python list:', time.time()-s)


s = time.time()
%time A1+A2
#print('numpy array:', time.time()-s)
```
